```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        else: 
            break
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()  
## error bcoz alreday appliead tha ye fn load mat le hm saala break dal rhe the twji mein saala
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

# SO ab hamare pas total 4806 movie vectors hai and ab hame har 4806 movie ka har doosri movie ke sath distance calculate krni hai.

- Distance jitna jyada similarity utna kam hoga.
- Yha pe we will not calculate eucledian distance but instead we'll calculate cosine distance.
- Cosine distance is basically angle b/w 2 vectors and agar ye angle bhut chota hai to distance bhut kam and similarity jyda
- Bcoz jb bhi hm high dim data se deal krte hai to eucledian distance utna fail krta bcoz it is not a relaible measure.
- cos_dist(x1,x2) = cos(theta)
- Niche we are using cos similarity and it lies b/w 0 and 1 i.e, 1 if they same and 0 if they are very dissimalar i.e, 180 degree.

## STEPS TO EXTRACT OUTPUT
- Sabse pahle mai given movie ki index fetch krunga.
- Ab mujhe is movie ki distances niklne hai and ye kya hoga similarity matrix ki wo particular movie ka indesx do.


```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    movie_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
    
```
